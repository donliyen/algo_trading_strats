```python
import pandas as pd
import pandas_ta as ta
import mplfinance as mpl
```

```python
df = pd.read_csv(r'C:\Users\zig53\vscode jupyther\Datasets\BTC-USD_6years.csv')

df.reset_index()

df['datetime'] = pd.to_datetime(df['datetime'])

df.index = df['datetime']
 
df.head()
```

```python
df['mean'] = df['close'].rolling(window = 120).mean()
df['std'] = df['close'].rolling(window = 120).std()
```

```python
df['Upper Band'] = df['mean'] + 2 * df['std']
df['Lower_Band'] = df['mean'] - 2 * df['std']
```

```python
df2 = df.loc[: , ['close' , 'mean','Upper Band','Lower_Band'] ]
```

```python
df2.plot(figsize = (20,12))
```
